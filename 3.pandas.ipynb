```python
#import pandas

import pandas as pd 
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
# set Name column as index

df.set_index("Name",inplace=True)
```

```python
df.head()
```

```python
#reset index 

df.reset_index(inplace=True)
```

```python
# missing value in each columns 

df.isnull().sum()
```

```python
#it delete null rows , ny default it delete rows

df.dropna()
```

```python
#it will fill the value where value is null

df.fillna("pratik")
```

```python
#it will delete those columns which contains null value , axis=1

df.dropna(axis=1)
```

```python
df.head()
```

```python
#performing groupby operation 

a  = df.groupby('Survived')
```

```python
a.mean()
```
