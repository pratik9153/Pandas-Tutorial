```python
data1 = pd.DataFrame({'key1' : [1,2,3,4,5,6],
                     'key2' : [4,5,6,7,8,9],
                     'key3' : [3,4,5,7,8,9]
})
```

```python
data2 = pd.DataFrame ({'key1' : [1,24,3,41,52,6],
                     'key5'   : [4,57,6,78,81,9],
                     'key6'   : [34,47,5,7,18,9]
})
```

```python
data1
```

```python
data2
```

```python
#merge operation

pd.merge(data1,data2)
```

```python
#merge operation based on left

pd.merge(data1,data2,how='left')
```

```python
#merge opeartion based on right

pd.merge(data1,data2,how='right')
```

```python
#merge opeartion based on outer

pd.merge(data1,data2,how='outer')
```

```python
#merge operation based on cross

pd.merge(data1,data2,how='cross')
```

## Join

```python
data1 = pd.DataFrame({'key1' : [1,2,3,4,5,6],
                     'key2' : [4,5,6,7,8,9],
                     'key3' : [3,4,5,7,8,9]},
                    index =  ['a','b','c','d','e','f']               
                    )
```

```python
data2 = pd.DataFrame ({'key4' : [1,24,3,41,52,6],
                     'key5'   : [4,57,6,78,81,9],
                     'key6'   : [34,47,5,7,18,9]},
                     index = ['a','b','i','j','k','l']
                     )
```

```python
data1
```

```python
data2
```

```python
#performing join opeartion 

data1.join(data2)
```

```python
#right join

data1.join(data2,how = 'right')
```

```python
#inner join 

data1.join(data2,how = 'inner')
```

```python
#outer join

data1.join(data2,how='outer')
```
